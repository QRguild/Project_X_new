```python
def get_trade_data(df, strike, option):
    """Extract relevant trade data for a specific strike and option."""
    matching_rows = df[(df['strike'] == strike) & (df['option'] == option)]
    return matching_rows

def calculate_price_levels(matching_row):
    """Calculate buy/sell price based on trading conditions."""
    buy_price = None
    is_buy = None
    
    if matching_row.BuyAvgPrice < matching_row.SellAvgPrice:
        if matching_row.MarkToMarket > 0:
            buy_price = round(matching_row.BuyAvgPrice / 0.05) * 0.05
            is_buy = True
        else:
            buy_price = round(matching_row.SellAvgPrice / 0.05) * 0.05
            is_buy = False
    elif matching_row.BuyAvgPrice > matching_row.SellAvgPrice:
        if matching_row.MarkToMarket < 0:
            buy_price = round(matching_row.BuyAvgPrice / 0.05) * 0.05
            is_buy = True
        else:
            buy_price = round(matching_row.SellAvgPrice / 0.05) * 0.05
            is_buy = False
            
    return buy_price, is_buy

def process_active_orders(matching_row, buy_price):
    """Process orders that are already active."""
    if matching_row.BuyAvgPrice < matching_row.SellAvgPrice:
        if matching_row.MarkToMarket > 0:
            return round(matching_row.SellAvgPrice / 0.05) * 0.05
        return round(matching_row.BuyAvgPrice / 0.05) * 0.05
    elif matching_row.BuyAvgPrice > matching_row.SellAvgPrice:
        if matching_row.MarkToMarket < 0:
            return round(matching_row.SellAvgPrice / 0.05) * 0.05
        return round(matching_row.BuyAvgPrice / 0.05) * 0.05
    return buy_price

def update_timestamps(timestamps, tradingsymbol, tick, buy_price, is_buy):
    """Update the timestamps dictionary with trade information."""
    if is_buy:
        timestamps[tradingsymbol] = {
            'buy_timestamp': tick.exchange_timestamp,
            'BuyAvgPrice': buy_price,
            'sell_timestamp': None,
            'SellAvgPrice': None
        }
    else:
        timestamps[tradingsymbol] = {
            'buy_timestamp': None,
            'BuyAvgPrice': None,
            'sell_timestamp': tick.exchange_timestamp,
            'SellAvgPrice': buy_price
        }

def get_trade_timestamps(data, path2):
    """Main function to process trade timestamps."""
    df = pd.read_csv(path2)
    df['strike'] = df['TradingSymbol'].str.extract(r'(\d+)$')
    df['option'] = df['TradingSymbol'].str.extract(r'\s(CE|PE)\s')
    
    timestamps = {}
    active_orders = []
    
    for tick in data.itertuples():
        option = tick.tradingsymbol[-2:]
        strike = str(int(tick.strike))
        matching_rows = get_trade_data(df, strike, option)

        buy_price = None
        for matching_row in matching_rows.itertuples():
            if tick.tradingsymbol not in active_orders:
                buy_price, is_buy = calculate_price_levels(matching_row)
            else:
                buy_price = process_active_orders(matching_row, buy_price)

            if buy_price is not None:
                price_lower = buy_price * 0.999
                price_upper = buy_price * 1.001

                if price_lower <= tick.last_price <= price_upper:
                    if tick.tradingsymbol not in active_orders:
                        active_orders.append(tick.tradingsymbol)
                        update_timestamps(timestamps, tick.tradingsymbol, tick, buy_price, is_buy)
                    else:
                        if timestamps[tick.tradingsymbol]['buy_timestamp'] is not None:
                            timestamps[tick.tradingsymbol]['SellAvgPrice'] = buy_price
                            timestamps[tick.tradingsymbol]['sell_timestamp'] = tick.exchange_timestamp
                        elif timestamps[tick.tradingsymbol]['sell_timestamp'] is not None:
                            timestamps[tick.tradingsymbol]['BuyAvgPrice'] = buy_price
                            timestamps[tick.tradingsymbol]['buy_timestamp'] = tick.exchange_timestamp
                        active_orders.remove(tick.tradingsymbol)

    return timestamps
```
