```python
import matplotlib.pyplot as plt
import pandas as pd
import re

df = pd.read_csv('log.csv')

# Extract timestamp column (assuming it's the first column or named 'Time')
timestamp_col = [col for col in df.columns if 'Time' in col or 'time' in col][0]
df['timestamp'] = df[timestamp_col]

df = df.drop(df.columns[[0, 1, 2]], axis=1)
df

# df.drop(df.columns[1], inplace=True)
# Function to extract the first float from a string
def extract_float(s):
    match = re.search(r"[-+]?\d*\.\d+|\d+", str(s))
    return float(match.group()) if match else None

# Extract float values for all columns except timestamp
for col in df.columns:
    if col != 'timestamp':
        df[col] = df[col].apply(extract_float)

# Drop unwanted columns and keep timestamp + numeric columns
cols_to_plot = [col for col in df.columns if col != timestamp_col and col != 'timestamp']
# df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
# df = df.dropna(subset=['timestamp'])

for i in range(len(df)):
    # df['timestamp'].iloc[i] = pd.to_datetime(df['timestamp'].iloc[i][6:])
    df.loc[i, 'timestamp'] = pd.to_datetime(df['timestamp'].iloc[i][6:])
    # print(pd.to_datetime(df['timestamp'].iloc[0][6:]))
# Plot each numeric column against timestamp

cols_to_plot = [col for col in df.columns if col != 'timestamp']
plt.figure(figsize=(12, 6))
for col in cols_to_plot:
    plt.plot(df['timestamp'], df[col], label=col)
plt.legend()
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title('Log Values Over Time')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import re

df = pd.read_csv('/Users/sushrutkagde/Documents/Project_X/2025-06-05_log.csv')

# Extract timestamp column (assuming it's the first column or named 'Time')
timestamp_col = [col for col in df.columns if 'Time' in col or 'time' in col][0]
df['timestamp'] = df[timestamp_col]

df = df.drop(df.columns[[0, 1, 2]], axis=1)
df

# df.drop(df.columns[1], inplace=True)
# Function to extract the first float from a string
def extract_float(s):
    match = re.search(r"[-+]?\d*\.\d+|\d+", str(s))
    return float(match.group()) if match else None

# Extract float values for all columns except timestamp
for col in df.columns:
    if col != 'timestamp':
        df[col] = df[col].apply(extract_float)

# Drop unwanted columns and keep timestamp + numeric columns
cols_to_plot = [col for col in df.columns if col != timestamp_col and col != 'timestamp']
# df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
# df = df.dropna(subset=['timestamp'])

for i in range(len(df)):
    # df['timestamp'].iloc[i] = pd.to_datetime(df['timestamp'].iloc[i][6:])
    df.loc[i, 'timestamp'] = pd.to_datetime(df['timestamp'].iloc[i][6:])
    # print(pd.to_datetime(df['timestamp'].iloc[0][6:]))
# Plot each numeric column against timestamp

cols_to_plot = [col for col in df.columns if col != 'timestamp']
plt.figure(figsize=(20, 9))
# for col in cols_to_plot:
#     plt.plot(df['timestamp'], df[col], label=col)
bps_threshold = 5
beta_adj_bps_threshold = 10
plt.legend()
plt.xlabel('Timestamp')
plt.axhline(y=bps_threshold, color='red', linestyle='--', label='BPS Threshold')
plt.axhline(y=beta_adj_bps_threshold, color='green', linestyle='--', label='Beta Adj BPS Threshold')

# Filter the dataframe to only include rows from 14:30 onwards
df_filtered = df[df['timestamp'] <= pd.to_datetime(df['timestamp'].iloc[0]).replace(hour=15, minute=30, second=0, microsecond=0)]

# Columns to plot (excluding ' Beta: 0')
cols_to_plot_filtered = [col for col in cols_to_plot if col.strip() != 'Beta: 0']

for col in cols_to_plot_filtered:
    plt.plot(df_filtered['timestamp'], df_filtered[col], label=col)

plt.axhline(y=-(bps_threshold), color='red', linestyle='--', label='BPS Threshold')
plt.axhline(y=-(beta_adj_bps_threshold), color='green', linestyle='--', label='Beta Adj BPS Threshold')
plt.legend()
plt.ylabel('Value')
plt.title('Log Values Over Time')
plt.ylim(-30, 30)
plt.tight_layout()
plt.show()
```
